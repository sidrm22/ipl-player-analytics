```python
import pandas as pd
```

```python
matches=pd.read_csv("matches.csv")
```

```python
matches.isnull().sum()  #See what's missing 
```

```python
matches['city'].fillna('Unknown', inplace=True) #Fill whats missing
```

```python
matches['city'].isnull().sum()
```

```python
matches["player_of_match"].fillna("Unknown",inplace=True)
```

```python
matches['city'].isnull().sum()
```

```python
matches.isnull().sum()
```

```python
name_map = {
    'Delhi Daredevils': 'Delhi Capitals',
    'Kings XI Punjab': 'Punjab Kings',
    'Royal Challengers Bangalore': 'Royal Challengers Bengaluru',
    'Deccan Chargers': 'Sunrisers Hyderabad',
    'Rising Pune Supergiant': 'Lucknow Super Giants',
    'Rising Pune Supergiants': 'Lucknow Super Giants',
    'Gujarat Lions': 'Gujarat Titans'
}
```

```python
matches['team1'] = matches['team1'].replace(name_map)
matches['team2'] = matches['team2'].replace(name_map)
```

```python
matches['winner'] = matches['winner'].replace(name_map)
matches['toss_winner'] = matches['toss_winner'].replace(name_map)
```

```python
for team in name_map.keys():
    print(team, (pd.concat([
        matches['team1'],
        matches['team2'],
        matches['winner'],
        matches['toss_winner']
    ]) == team).sum())
```
